```python
import os
import pandas as pd

def read_csv_files_from_folder(folder_path):
    all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]
    df_list = []
    
    for file in all_files:
        df = pd.read_csv(file)
        df_list.append(df)
        
    return pd.concat(df_list, ignore_index=True)

path_to_panda_1_csv = '/workspaces/wiggle-face/data/panda_each_video/panda1_features_complete'
path_to_panda_2_csv = '/workspaces/wiggle-face/data/panda_each_video/panda2_features_complete'
path_to_panda_3_csv = '/workspaces/wiggle-face/data/panda_each_video/panda3_features_complete'

df_panda1 = read_csv_files_from_folder(path_to_panda_1_csv)
df_panda2 = read_csv_files_from_folder(path_to_panda_2_csv)
df_panda3 = read_csv_files_from_folder(path_to_panda_3_csv)
df_all = pd.concat([df_panda1, df_panda2, df_panda3], ignore_index=True)

print(df_all)

output_path = '/workspaces/wiggle-face/data/panda_each_video/all_panda_features_updated.csv'
df_all.to_csv(output_path, index=False)
print(f"Combined dataframe saved to {output_path}")
```
